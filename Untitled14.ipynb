```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# creating a simple list of hostel water tank levels
maximum_water_level =500
minimum_water_level =0
#stimulate water level for 24 hours
levels = np.random.randint(low=minimum_water_level, high=maximum_water_level + 1, size=24)
hours= range(24)
df = pd.DataFrame({'level': levels}, index=hours)
# for calculating the "usage"(the difference between one hour and the next hour)
# here we are using.diff() to see how much the level of water changed
df['usage'] = df['level'].diff()
#we are looking only at usage > 0(ignoring the hours when the level of water is full)
df = df[df['usage'] > 0]
print(f"average usage: {df['usage'].mean():.2f}")
#creating the line chart
plt.plot(df.index, df['usage'])
plt.xlabel('hours')
plt.ylabel('usage')
plt.title('water usage')
plt.show()
```
